```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
DATA_PATH_SINHVIEN_RAW = '/content/drive/MyDrive/Nhóm 6 - DS317.P11/ĐỒ ÁN MÔN HỌC/2.dataset_raw/01.sinhvien_raw.csv'
DATA_PATH_THI_SINH_RAW = '/content/drive/MyDrive/Nhóm 6 - DS317.P11/ĐỒ ÁN MÔN HỌC/2.dataset_raw/05.ThiSinh.xlsx'
```

```python
df_sinhvien = pd.read_csv(DATA_PATH_SINHVIEN_RAW)
df_thi_sinh = pd.read_excel(DATA_PATH_THI_SINH_RAW)
```

```python
# Create a new column to group all other categories into 'OTHER'
df_thi_sinh['encoded_column'] =df_thi_sinh['dien_tt'].apply(lambda x: x if x in ['THPT', 'ĐGNL'] else 'OTHER')

# Perform one-hot encoding
encoded_columns = pd.get_dummies(df_thi_sinh['encoded_column'], prefix='', prefix_sep='')

# Add the encoded columns to the original DataFrame
df = pd.concat([df_thi_sinh, encoded_columns], axis=1)
```

```python
# Replace True with 1 and False with 0
df = df.replace({True: 1, False: 0})
```

```python
df.info()
```

```python
df.drop(columns = ['dien_tt','lop12_matinh','lop12_matruong','TEN_TRUONG','encoded_column'],inplace = True)
```

```python
df.isnull().sum()
```

```python
df_tabular = pd.merge(df, df_sinhvien, on='mssv', how='inner')
```

```python
df_tabular.drop(columns = ['id','namsinh','gioitinh','khoahoc','tinhtrang','lopsh','name_cn2'],inplace = True)
```

```python
import pandas as pd

# Assume `df` is your DataFrame
columns = list(df_tabular.columns)

# Extract noisinh columns (assuming they are named in a pattern)
noisinh_columns = [col for col in columns if col.startswith("noisinh_")]

# Remove noisinh columns from the original order
remaining_columns = [col for col in columns if col not in noisinh_columns]

# Determine the position where noisinh columns should be inserted
target_position = 36

# Insert noisinh columns at the target position
new_columns_order = (
    remaining_columns[:target_position] +
    noisinh_columns +
    remaining_columns[target_position:]
)

# Reorder the DataFrame columns
df_tabular = df_tabular[new_columns_order]

# Check the new column order
print(df_tabular.columns)
```

```python
df_tabular.info()
```

```python
# Identify the noisinh columns
noisinh_columns = [col for col in df_tabular.columns if col.startswith("noisinh_đc")]

# Rename noisinh columns to noisinh_0, noisinh_1, etc.
rename_dict = {col: f"noisinh_{i}" for i, col in enumerate(noisinh_columns)}

# Apply renaming
df_tabular.rename(columns=rename_dict, inplace=True)

# Verify changes
print(df_tabular.columns)
```
